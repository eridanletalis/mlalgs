1. К алгоритму kNN, реализованному на уроке, реализовать добавление весов для соседей по любому из показанных на уроке принципов.
Модифицированная функция после примеров, взятых из методички.

```python
import numpy as np
from sklearn import model_selection
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
X, y = load_iris(return_X_y=True)

# Для наглядности возьмем только первые два признака (всего в датасете их 4)
X = X[:, :2]
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
cmap = ListedColormap(['red', 'green', 'blue'])
plt.figure(figsize=(7, 7))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap)
```

```python
def e_metrics(x1, x2):
    
    distance = 0
    for i in range(len(x1)):
        distance += np.square(x1[i] - x2[i])
    
    return np.sqrt(distance)
```

```python
def knn(x_train, y_train, x_test, k):
    
    answers = []
    for x in x_test:
        test_distances = []
            
        for i in range(len(x_train)):
            
            # расчет расстояния от классифицируемого объекта до
            # объекта обучающей выборки
            distance = e_metrics(x, x_train[i])
            
            # Записываем в список значение расстояния и ответа на объекте обучающей выборки
            test_distances.append((distance, y_train[i]))
        
        # создаем словарь со всеми возможными классами
        classes = {class_item: 0 for class_item in set(y_train)}
        
        # Сортируем список и среди первых k элементов подсчитаем частоту появления разных классов
        for d in sorted(test_distances)[0:k]:
            classes[d[1]] += 1
            
        # Записываем в список ответов наиболее часто встречающийся класс
        answers.append(sorted(classes, key=classes.get)[-1])
    return answers
```

```python
def accuracy(pred, y):
    return (sum(pred == y) / len(y))
```

```python
k = 1

y_pred = knn(X_train, y_train, X_test, k)

print(f'Точность алгоритма при k = {k}: {accuracy(y_pred, y_test):.3f}')
```

```python
def get_graph(X_train, y_train, k):
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA','#00AAFF'])

    h = .02

    # Расчет пределов графика
    x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
    y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Получим предсказания для всех точек
    Z = knn(X_train, y_train, np.c_[xx.ravel(), yy.ravel()], k)

    # Построим график
    Z = np.array(Z).reshape(xx.shape)
    plt.figure(figsize=(7,7))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Добавим на график обучающую выборку
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cmap)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(f"Трехклассовая kNN классификация при k = {k}")
    plt.show()
```

```python
get_graph(X_train, y_train, k)
```

```python
# Реализуется метод взвешивания в зависимости от порядкового номера 

def knn_w(x_train, y_train, x_test, k):
    answers = []
    for x in x_test:
        test_distances = []

        for i in range(len(x_train)):
            # расчет расстояния от классифицируемого объекта до
            # объекта обучающей выборки
            distance = e_metrics(x, x_train[i])

            # Записываем в список значение расстояния и ответа на объекте обучающей выборки
            test_distances.append((distance, y_train[i]))

        # создаем словарь со всеми возможными классами
        classes = {class_item: 0 for class_item in set(y_train)}

        # Сортируем список и среди первых k элементов подсчитаем частоту появления разных классов

        # Сначала отсортировали, потом посчитали

        for i, d in enumerate(sorted(test_distances)[0:k]):
            classes[d[1]] += (k + 1 - i) / k  # /1

        pred = int(sorted(classes, key=classes.get)[-1])
        # Записываем в список ответов наиболее часто встречающийся класс
        answers.append(pred)
    return answers
```

```python
k = 10

y_pred = knn_w(X_train, y_train, X_test, k)

print(f'Точность алгоритма при k = {k}: {accuracy(y_pred, y_test):.3f}')
```

```python
get_graph(X_train, y_train, k)
```

Видно, что при добавлении веса, обучение становится немного лучше. На графиках также видно, что области, на которые разделяется плоскость, становятся более "кучными", локализуясь, преимущественно, в одном месте. Возможно, есть такие примеры, на которых добавление веса будет сильно влиять на результат, но текущий игрушечный набор данных - не тот случай. На графиках ниже информация, из-за которой сделан такой вывод. Графики наложены друг на друга.

```python
knn_w_acc = []
knn_acc = []
for k in range(1,len(X_train) - 1):

    y_pred = knn_w(X_train, y_train, X_test, k)
    knn_w_acc.append([k, accuracy(y_pred, y_test)])
    y_pred = knn_w(X_train, y_train, X_test, k)
    knn_acc.append([k, accuracy(y_pred, y_test)])
    
knn_w_acc = np.array(knn_w_acc)
knn_acc = np.array(knn_acc)
```

```python
plt.plot(knn_acc[:,0], knn_acc[:,1], label="Accuracy без взвешивания")
plt.plot(knn_w_acc[:,0], knn_w_acc[:,1], alpha=0.5, label="Accuracy со взвешиванием")
plt.xlabel("Количество точек knn")
plt.ylabel("Accuracy")
plt.legend()
```
